## Pdfminer

```python
def parse_doc(doc, mode='return_as_string'):
    numPages = 0
    rsrcmgr = PDFResourceManager()
    laparams = LAParams()

    if mode=='return_as_list':
        lines = []
        device = PDFPageAggregator(rsrcmgr, laparams=laparams)
    else:
        string = StringIO()
        device = TextConverter(rsrcmgr, string, laparams=laparams)
    interpreter = PDFPageInterpreter(rsrcmgr, device)
 
    for page in PDFPage.create_pages(doc):
        numPages += 1
        interpreter.process_page(page)

        if mode=='return_as_list':
            layout = device.get_result()
            for element in layout:
                if isinstance(element, LTTextBoxHorizontal):
                    lines.extend(element.get_text().splitlines())
     
    if mode=='return_as_list':
        result = lines
    else:
        result = string.getvalue()
        string.close()
        
    device.close()
    
    return result, numPages
```

```python
with open(input_path, "rb") as f:
    parser = PDFParser(f)
    doc = PDFDocument(parser)     
    content, numPages = parse_doc(doc, mode='return_as_string') 
    print("Number of pages =", numPages)
    print(content)
```

# Text Summarization

```python
with open(input_path, "rb") as f:
    parser = PDFParser(f)
    doc = PDFDocument(parser)     
    content, numPages = parse_doc(doc, mode='return_as_string') 
```

```python
content_cleaned, _, _ = content.partition('\nReferences\n')

content_cleaned = content_cleaned.replace('-\n', '')
content_cleaned = content_cleaned.replace('\n\n', '. ')

#content_cleaned = re.sub(r'[^a-zA-Z.]', ' ', content_cleaned)
content_cleaned = re.sub(r'\s+', ' ', content_cleaned)

content_cleaned = content_cleaned.replace(' .', '.')
content_cleaned = re.sub(r'\.+', '.', content_cleaned)

content_cleaned = content_cleaned.lower()

print(content_cleaned)
```

## Gensim

```python
summary = summarize(content_cleaned, ratio=0.05)
print(summary)
```

## Textrank

```python
word_emb = {}
with open(glove_path, encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        word_emb[word] = coefs
print(len(word_emb))
```

```python
sentences = nltk.tokenize.sent_tokenize(content_cleaned)
print(len(sentences))
print(sentences[:5])
print(sentences[-5:])
```

```python
sentence_vectors = []
for s in sentences:
    if len(s)!=0:
        v = sum([word_emb.get(w, np.zeros((glove_size,))) for w in s.split()]) / (len(s.split())+0.001)
    else:
        v = np.zeros((glove_size,))
    sentence_vectors.append(v)
print(len(sentence_vectors))
```

```python
sim_mat = np.zeros([len(sentences), len(sentences)])
for i in range(len(sentences)):
    for j in range(len(sentences)):
        if i != j:
            sim_mat[i][j] = cosine_similarity(sentence_vectors[i].reshape(1,glove_size), 
                                              sentence_vectors[j].reshape(1,glove_size))[0,0]
print(sim_mat)
```

```python
nx_graph = nx.convert_matrix.from_numpy_matrix(sim_mat)
scores = nx.pagerank(nx_graph)

ranked_sentences = sorted(((scores[i], s) for i, s in enumerate(sentences)), reverse=True)
for i in range(10):
    print(ranked_sentences[i][1])
```
